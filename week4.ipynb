return np.array([float(item) for item in text.rstrip().split("\t")], dtype=float) แปลงค่าเป็นทศนิยม

```python

import numpy as np
def tab2float(text):
    return np.array([float(item) for item in text.rstrip().split("\t")], dtype=float)

ETavg = tab2float("0.815	0.631	0.798	0.832	0.887")
LRavg = tab2float("0.756	0.515	0.744	0.768	0.818")
MLPavg = tab2float("0.788	0.579	0.779	0.797	0.865")
RFavg = tab2float("0.815	0.630	0.813	0.817	0.898")
SVMavg = tab2float("0.807	0.614	0.815	0.799	0.893")

ETtest = tab2float("0.808	0.618	0.775	0.841	0.889")
LRtest = tab2float("0.752	0.509	0.716	0.788	0.808")
MLPtest = tab2float("0.797	0.598	0.753	0.841	0.871")
RFtest = tab2float("0.814	0.628	0.792	0.835	0.904")
SVMtest = tab2float("0.805	0.610	0.783	0.826	0.899")
```

```python
ETavg
```

```python
import os
labels = ['ACC', 'MCC', 'Sn', 'Sp', 'AUC']

plt.rcParams.update({'font.size': 14})
x = np.arange(len(labels))  # the label locations
width = 0.16  # the width of the bars
plt.figure(figsize=(2*8.27,11.69/2))

ax1 = plt.subplot2grid((1,2), (0,0))
ax1.bar(x - 2*width, ETavg, width, label='ET')
ax1.bar(x - width, LRavg, width, label='LR')
ax1.bar(x, MLPavg, width, label='MLP')
ax1.bar(x + width, RFavg, width, label='RF')
ax1.bar(x + 2*width, SVMavg, width, label='SVM')
ax1.text(-0.1, 1.0, 'A', transform=ax1.transAxes,  fontsize=22, weight='bold')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax1.set_ylim([0,1.05])
ax1.set_ylabel('Average Performances', fontsize=16)
ax1.set_xticks(x)
ax1.set_xticklabels(labels, fontsize=16, rotation=45)

ax2 = plt.subplot2grid((1,2), (0,1))
ax2.bar(x - 2*width, ETtest, width, label='ET')
ax2.bar(x - width, LRtest, width, label='LR')
ax2.bar(x, MLPtest, width, label='MLP')
ax2.bar(x + width, RFtest, width, label='RF')
ax2.bar(x + 2*width, SVMtest, width, label='SVM')
ax2.text(-0.1, 1.0, 'B', transform=ax2.transAxes,  fontsize=22, weight='bold')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax2.set_ylim([0,1.05])
ax2.set_ylabel('Average Performances', fontsize=16)
ax2.set_xticks(x)
ax2.set_xticklabels(labels, fontsize=16, rotation=45)
ax2.legend(ncol=3,handleheight=1, labelspacing=0.05, markerscale = 0.5, fontsize=14)

plt.tight_layout(pad=0.8, w_pad=3, h_pad=1.0)
plt.savefig('%s/Figure1.pdf'%(os.getcwd()), format='pdf', dpi=450)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(15,6))

# bar chart example
accounts = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
steem = [1100, 500, 1200, 2300]

plt.subplot(1,2,1)
plt.title('Bar chart')
plt.ylabel('Steem')
plt.bar(accounts, steem)

# pie chart example
labels = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
perc = [steem[0]/sum(steem), # calculate percentages from the amount of steem
        steem[1]/sum(steem), 
        steem[2]/sum(steem), 
        steem[3]/sum(steem)]

explode = (0, 0, 0.5, 0)     # "explode" the 3rd slice ('Scipio')

plt.subplot(1,2,2)
plt.title('Pie chart')
plt.axis('equal')            # Make sure the pie is drawn as a circle, X and Y axes equally sized
plt.pie(perc, labels=labels, autopct='%1.2f%%', 
        startangle=0, explode=explode)

plt.show()
```
