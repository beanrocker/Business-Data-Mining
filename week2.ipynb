```python
a = [item*2 for item in L]
a
```

```python
a = [item*2 for item in L if item is not'B']
a
```

def = define function

```python
def summation(L):
    z = 0
    for item in L:
        z = z + item
    return z
summation([1,2,3,4])
```

```python
def summation(L):
    z = ""
    for item in L:
        z = z + item
    return z
summation(["A","B","C"])
```

```python
def multi(start,end):
    for i in range(start,end):
        for j in range(12):
            print(str(i) + "x" + str(j+1)+" = "+str(i*j+1))
        print("\n")
multi(1,5)
```
