```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
#from sklearn.feature_selection import SelectKBest
#from sklearn.feature_selection import f_classif
from sklearn.pipeline import Pipeline
sm = SMOTE(random_state=1)
X_res, y_res = sm.fit_resample(X_train, y_train)


clf = Pipeline([
  ('scaler', MinMaxScaler(feature_range=(0, 1))),
  #('feature_selection', SelectKBest(f_classif)),
  ('classification', RandomForestClassifier(random_state=0))
])
```

```python
params = {
    #'feature_selection__k' : [1,4,7,10],
    'classification__n_estimators' : [10,20,100,150,200]
}
best_clf = GridSearchCV(clf, params, cv=10)
best_clf.fit(X_res, y_res)
```

```python
acc = best_clf.best_score_
print("10CV Training Accuracy : "+str(acc))
```

```python
best_clf.best_params_
```

```python
y_predict = best_clf.predict(X_test)
acc = sum(y_predict == y_test)/len(y_test)
print("Test accuracy : "+str(acc))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
target_names = ['not-Churn', 'Churn']
C = confusion_matrix(y_test,y_predict) 
C = C / C.astype(np.float).sum(axis=1)*100
sns.heatmap(C, annot=True, fmt="f",cmap="GnBu",xticklabels=target_names, yticklabels=target_names)
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict, target_names=target_names))
```
